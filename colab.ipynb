```python
# Текст, который будет озвучен
text_to_speech = """
Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do.
Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it
"""
```

```python
import http.client
import json
import base64
```

```python
# Устанавливаем соединение с API
conn = http.client.HTTPSConnection("text-to-speech-neural-google.p.rapidapi.com")
```

```python
# Подготавливаем данные для запроса
payload = json.dumps({
    "audioFormat": "ogg",
    "paragraphChunks": [text_to_speech],
    "voiceParams": {
        "name": "Wavenet-B",
        "engine": "google",
        "languageCode": "en-IN"
    }
})
```

```python
headers = {
    'x-rapidapi-key': "ae93df2ff8mshc817b2e69095a4cp18f652jsn862a14f962a4",
    'x-rapidapi-host': "text-to-speech-neural-google.p.rapidapi.com",
    'Content-Type': "application/json"
}
```

```python
# Отправляем запрос к API
conn.request("POST", "/generateAudioFiles", payload, headers)
```

```python
# Получаем ответ от сервера
res = conn.getresponse()
data = res.read()
```

```python
# Декодируем данные из JSON
response_json = json.loads(data.decode("utf-8"))
```

```python
# Получаем закодированный аудиопоток в Base64
audio_base64 = response_json["audioStream"]
```

```python
# Декодируем аудиопоток из Base64
audio_data = base64.b64decode(audio_base64)
```

```python
# Сохраняем аудио в формате OGG
with open("output_audio_alisa.ogg", "wb") as file:
    file.write(audio_data)
```

```python
print("Аудиофайл сохранен как output_audio_alisa.ogg")
```

```python
from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_voice
```

```python
# Инициализация модели
tts = TextToSpeech()
```

```python
# Имя выбранного голоса
voice_name = "geralt"
```

```python
# Загрузка голосовых файлов
voice_samples, conditioning_latents = load_voice(voice_name)
```

```python
# Генерация аудио
gen_audio = tts.tts_with_preset(text_to_speech, voice_samples=voice_samples, conditioning_latents=conditioning_latents, preset="ultra_fast")
```

```python
import numpy as np
from scipy.io.wavfile import write
```

```python
# Преобразуем Tensor в numpy массив
audio_np = gen_audio.cpu().numpy()
```

```python
# Убедитесь, что данные имеют правильный тип (например, float32 или int16)
audio_np = np.int16(audio_np * 32767)
```

```python
# Сохраняем файл как WAV
write("output_audio_alisa_geralt.wav", 22050, audio_np)  # 22050 — это частота дискретизации
```
